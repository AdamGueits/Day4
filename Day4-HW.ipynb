### Here is Example Code for a simple Histogram

```python
fig = px.histogram(DF,
                   nbins=10,
                   x='loan_amount',
                   color_discrete_sequence=['lightseagreen'])

fig.update_layout(bargap=0.1,
                  title='Histogram of Loan Ammounts.',
                  title_x=0.5)
fig.show()
```

### Here is example code for a histogram with a box plot included.

```python
fig = px.histogram(DF,
                   nbins=9,
                   x='loan_amount',
                   color='homeownership',
                   opacity=0.5,
                   color_discrete_map={'MORTGAGE': 'darkviolet', 
                                       'RENT': 'deeppink', 
                                       'OWN': 'darkturquoise'},
                   marginal="violin"
                   )

fig.update_layout(bargap=0.1,
                  title='Histogram of Loan Ammounts.',
                  title_x=0.5,
                  xaxis={'range':[-1000, 46000]},
                  xaxis_title="Loan Amount ($)",
                  yaxis_title="Frequency",
                  legend_title="Homeownership",
                  autosize=False,
                  width=800,
                  height=600)
fig.show()
```

### Here is example code for a bar plot

```python
fig = px.bar(DF,
            x='homeownership',
            color_discrete_sequence=['lightseagreen'])
fig.update_traces(dict(marker_line_width=0))

fig.show()
```

**Q7** Can you figure out how to add x labels, y labels, and a title to this graph?

```python
# Your code here
fig = px.bar(DF,
            x='homeownership',
            color_discrete_sequence=['lightseagreen'],)
fig.update_traces(dict(marker_line_width=0))
fig.update_layout(bargap=0.1,
                  title='Histogram of Homeownership.',
                  title_x=0.5,xaxis_title="Homeownership",yaxis_title="Count")

fig.show()
```
